```python
from scipy.stats import skew
```

```python
from scipy.stats import kurtosis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in data:
  print(col)
  print(skew(data[col]))

  plt.figure()
  sns.distplot(data[col])
  plt.show()
```

```python
for col in data:
  print(col)
  print(kurtosis(data[col]))

  plt.figure()
  sns.distplot(data[col])
  plt.show()
```
