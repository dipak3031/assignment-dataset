```python
import pandas as pd
```

```python
data = pd.read_csv("https://github.com/dipak3031/assignment-dataset/files/10089428/Q9_a.csv")
```

```python
data
```

```python
import scipy
```

```python
from scipy.stats import skew
```

```python
print(skew(data,axis=0,bias=True))
```

```python
from scipy.stats import kurtosis
```

```python
print(kurtosis(data,axis=0,bias=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in data:
  print(col)
  print(skew(data[col]))

  plt.figure()
  sns.distplot(data[col])
  plt.show()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
for col in data:
  print(col)
  print(kurtosis(data[col]))

  plt.figure()
  sns.distplot(data[col])
  plt.show()
```
