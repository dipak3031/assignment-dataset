```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# loading the dataset
data=pd.read_csv('https://github.com/dipak3031/assignment-dataset/files/10104106/LabTAT.csv')
data
```

```python
data.head()
```

```python
# Anova ftest statistics: stats.f_oneway(column-1,column-2,column-3,column-4)
p_value=stats.f_oneway(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])
p_value
```
