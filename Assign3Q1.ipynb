Assignment-03-Q1 (Hypothesis Testing)


```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#loading the dataset
df=pd.read_csv('https://github.com/dipak3031/assignment-dataset/files/10104104/Cutlets.csv')
df
```

```python
df.head()
```

```python
unit_a=pd.Series(df.iloc[:,0])
unit_a
```

```python
unit_b=pd.Series(df.iloc[:,1])
unit_b
```

```python
#here 2-sample so 2-tail ttest is applicable:   stats.ttest_ind(array1,array2)    
#ind -> independent samples
p_value=stats.ttest_ind(unit_a,unit_b)
p_value
```

```python
p_value[1]     #for 2-tail probability 
```
